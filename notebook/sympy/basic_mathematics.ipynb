```python
# Exponents and Logarithms

from math import log, isclose


# Definimos variables base
x = 2
a = 5
b = 2
m = 3
n = 4

# 1. Multiplication Property: x^m * x^n = x^(m+n)
left_mult = x**m * x**n
right_mult = x**(m + n)
assert isclose(left_mult, right_mult), "Multiplication property failed!"
# Logarithm: log(a * b) = log(a) + log(b)
log_mult = log(a * b)
log_mult_expected = log(a) + log(b)
assert isclose(log_mult, log_mult_expected), "Multiplication property failed!"

# 2. Division Property: x^m / x^n = x^(m-n)
left_div = x**m / x**n
right_div = x**(m - n)
assert isclose(left_div, right_div), "Division property failed!"
# Logarithm: log(a / b) = log(a) - log(b)
log_div = log(a / b)
log_div_expected = log(a) - log(b)
assert isclose(log_div, log_div_expected), "Division property failed!"

# 3. Exponentiation Property: (x^m)^n = x^(mn)
left_exp = (x**m)**n
right_exp = x**(m * n)
assert isclose(left_exp, right_exp), "Exponentiation property failed!"
# Logarithm: log(a^n) = n * log(a)
log_exp = log(a**n)
log_exp_expected = n * log(a)
assert isclose(log_exp, log_exp_expected), "Exponentiation property failed!"

# 4. Zero Exponent Property: x^0 = 1
zero_exp = x**0
assert zero_exp == 1, "Zero exponent property failed!"
# Logarithm: log(1) = 0
log_zero_exp = log(1)
assert log_zero_exp == 0, "Zero exponent property failed!"

# 5. Inverse Property: x^-1 = 1/x
left_exp_inverse = x**-1
right_exp_inverse = 1/x
assert isclose(left_exp_inverse, right_exp_inverse), "Inverse property failed!"
# Logarithm: log(x^-1) = -log(x)
left_log_inv = log(x**-1)
right_log_inv = -log(x)
assert isclose(left_log_inv, right_log_inv), "Inverse property failed!"
```

```python
# Euler’s Number
from math import exp  # Euler's number
from sympy import symbols, limit, oo

p = 100
r = .20
t = 2.0


n = 12
ci_n_annual = p * (1 + (r/n))**(n * t)

n = 365
ci_n_daily = p * (1 + (r/n))**(n * t)

n = 8760
ci_n_hour = p * (1 + (r/n))**(n * t)

n = 525600
ci_n_minute= p * (1 + (r/n))**(n * t)

print(ci_n_annual, ci_n_daily, ci_n_hour, ci_n_minute)

# Euler's number using limit definition

n_sym = symbols('n')

euler_expr = (1 + 1/n_sym)**n_sym
euler_10000000 = euler_expr.subs(n_sym, 10000000)
euler_limit = limit(euler_expr, n_sym, oo)

print(float(euler_limit), float(euler_10000000))  # 2.718281828459045
assert isclose(float(euler_limit), float(euler_10000000), rel_tol=1e-9), "Euler's number limit failed!"

ci_n_limit = limit(p * (1 + (r/n_sym))**(n_sym * t), n_sym, oo)
ci_euler = p * exp(r*t) # e^{x}
print(float(ci_n_limit), ci_euler)

assert isclose(ci_euler, float(ci_n_limit), rel_tol=1e-9), "Euler's compose interes!"
```

```python
# Limits

from sympy import symbols, limit, oo

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)

print(result)


n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

print(result) # E
print(result.evalf()) # 2.71828182845905
```

```python
# Derivatives

def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m


def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2) # prints 4.000010000000827
```

```python
# Partial Derivatives


from sympy import symbols, diff
from sympy.plotting import plot3d

# Declare x and y to SymPy
x,y = symbols('x y')

# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3

# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f) # prints 6*x**2
print(dy_f) # prints 9*y**2

# plot the function
plot3d(f)
```

```python
from sympy import symbols, limit

# "x" and step size "s"
x, s = symbols('x s')

# declare function
f = x**2

# slope between two points with gap "s"
# substitute into rise-over-run formula
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)

# substitute 2 for x
slope_2 = slope_f.subs(x, 2)

# calculate slope at x = 2
# infinitely approach step size _s_ to 0
slope = result = limit(slope_f, s, 0)
result = limit(slope_2, s, 0)

print(slope) # 2x
print(result) # 4
```

```python
# The chain rule

# This is the chain rule, which says that for a given function y (with input variable x) 
# composed into another function z (with input variable y), 
# we can find the derivative of z with respect to x by multiplying the two respective derivatives together:

from sympy import symbols, diff

x, y = symbols('x y')

_y = x**2 + 1
z = y**3 - 2

# derivative for first function
# need to underscore y to prevent variable clash

dy_dx = diff(_y)

# derivative for second function
dz_dy = diff(z)

# Calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain) # 6*x*(x**2 + 1)**2
print(dz_dx_no_chain) # 6*x*(x**2 + 1)**2
```

```python
# Integrals

# The opposite of a derivative is an integral, which finds the area under the curve for a given range
# we will be finding the areas under probability distributions. Although we will not use integrals directly, 
# and instead will use cumulative density functions that are already integrated, 

# Riemann Sums
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0

    for i in range(n):
        # Calcular el punto medio del subintervalo
        midpoint = a + (i + 0.5) * delta_x
        total_sum += f(midpoint) * delta_x

    return total_sum

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)
print(area)  # Debería imprimir un valor más preciso para el área bajo la curva

area = approximate_integral(a=0, b=1, n=1000, f=my_function)
print(area)  # Debería imprimir un valor más preciso para el área bajo la curva

area = approximate_integral(a=0, b=1, n=1_000_000, f=my_function)
print(area)  # Debería imprimir un valor más preciso para el área bajo la curva
```

```python
# Integrals

# ∫ x^n dx = (x^(n+1)) / (n+1) + C   , para n ≠ -1
# ∫ x^2 dx = (x^3)/3 + C

from sympy import symbols, integrate

# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f = x**2 + 1

# Calculate the integral of the function with respect to x
# for the area between x = 0 and 1
area = integrate(f, (x, 0, 1))

print(area) # prints 4/3 ≈ 1.33333333333333
```

```python
# The main idea follows much of what we did earlier: 
# pack rectangles under a curve and make them infinitely smaller until we approach the exact area. 
# But of course, the rectangles cannot have a width of 0…​they just have to keep getting closer to 0 
# without ever reaching 0. It is another case of using limits.

from sympy import symbols, Sum, limit, oo

# Declare variables to SymPy
x, i, n = symbols('x i n')

# Declare function and range
f = x**2 + 1
lower, upper = 0, 1

# Calculate width and each rectangle height at index "i"
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

# Iterate all "n" rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# Calculate the area by approaching the number
# of rectangles "n" to infinity
area = limit(n_rectangles, n, oo)

print(area) # prints 4/3
```
