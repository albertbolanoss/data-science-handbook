# rbf_kernel 

La funci√≥n rbf_kernel de sklearn.metrics.pairwise calcula la similitud Gaussiana (RBF - Radial Basis Function) entre cada par de elementos en dos conjuntos de datos. Se usa com√∫nmente en m√©todos de Machine Learning como SVMs y Kernel Ridge Regression.

La similitud se calcula con la funci√≥n de base radial (RBF):

K(x, y) = exp (- Œ≥ ‚à•x - y‚à•^2 )

x y ùë¶ son vectores de entrada (pueden ser matrices).
Œ≥ es un hiperpar√°metro que controla la escala de la similitud.
‚à•ùë•‚àíùë¶‚à•^2 es la distancia euclidiana al cuadrado entre los puntos.
exp es la funci√≥n exponencial.

üìå Cuanto m√°s cerca est√©n ùë• y y, mayor ser√° el valor de ùêæ(ùë•,ùë¶) K(x,y) (m√°s cercanos a 1).
üìå Si est√°n lejos, ùêæ(ùë•,ùë¶)K(x,y) se acercar√° a 0.



```python
import numpy as np
from sklearn.metrics.pairwise import rbf_kernel

# Conjunto de datos de edades
housing_median_age = np.array([[20], [25], [30], [35], [40], [45]])

# Punto de referencia: edad 35
reference = np.array([[35]])

# Calcular similitud RBF con gamma=0.1
gamma = 0.1
similarity = rbf_kernel(housing_median_age, reference, gamma=gamma)

# Mostrar resultados
for age, sim in zip(housing_median_age.flatten(), similarity.flatten()):
    print(f"Similitud entre {age} y 35: {sim:.4f}")
```

```python
from sklearn.svm import SVC
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np

# Generar datos no linealmente separables
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

# Entrenar un SVM con kernel RBF
svm_clf = SVC(kernel="rbf", gamma=0.5, C=1.0)
svm_clf.fit(X, y)

# Graficar la clasificaci√≥n
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm")
plt.title("Clasificaci√≥n con SVM y Kernel RBF")
plt.show()

X
```
